```python
df_te = df[['Chinese','Math','English']]
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]
def color_red(val):
    color = 'red' if val > 95.0 else 'black'
    return 'color: %s' % color
df_te.style.applymap(color_red).apply(highlight_max)
```
